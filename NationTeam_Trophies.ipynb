# National Team Preprocessing

```python
selected_columns = ["National Team", "National Status", "National Matches", "National Goals", "National Assists", "National Yellow Cards", "National Red Cards"]
filtered_data = data[selected_columns]

filtered_data
```

Есть строки с NaN, тк у некоторых футболистов вообще нет истории вызова в сборную (их не приглашали).
Поэтому NaN меняем на нули и меняем "National Status".

```python
data = data.fillna(0)

data['National Team'].replace({0: 'Not Invited'}, inplace=True)
data.loc[data['National Team'] == 'Not Invited', 'National Status'] = data['National Status'].replace(0, -1)

filtered_data = data[selected_columns]
filtered_data
```

*Update:* National Status

Теперь у нас есть **четыре статуса выступления за национальную сборную**:
- -1: игрок никогда не вызывался в сборную
-  0: футболист не является действующим игроком своей сборной
-  1: игрок выступает за молодежку сборной
-  2: футболист является действующим игроком своей сборной

```python
# 2: 'Brazil'|'Argentina'|'England'| 'France'| 'Belgium'| 'Netherlands'| 'Italy'| 'Portugal'| 'Spain'|'Croatia'

# 1: 'Uruguay'| 'USA'| 'Morocco'| 'Colombia'| 'Mexico'| 'Germany'| 'Japan'| 'Switzerland'| 'Denmark'| 'Senegal'
```

```python
# Словарь для соответствия сборных и их рейтинга
team_ranking = {
    'Brazil': 2, 'Argentina': 2, 'England': 2, 'France': 2, 'Belgium': 2, 'Netherlands': 2,
    'Italy': 2, 'Portugal': 2, 'Spain': 2, 'Croatia': 2,
    'Uruguay': 1, 'USA': 1, 'Morocco': 1, 'Colombia': 1, 'Mexico': 1, 'Germany': 1,
    'Japan': 1, 'Switzerland': 1, 'Denmark': 1, 'Senegal': 1
}

data['National Team Ranking'] = data['National Team'].map(team_ranking).fillna(0)

# Размещаем 'National Team Ranking' после 'National Team'
national_team_index = data.columns.get_loc("National Team")
data = data[list(data.columns[:national_team_index + 1]) + ['National Team Ranking'] + list(data.columns[national_team_index + 1:])]

data = data.iloc[:, :-1]
data
```

## trophies_in_recent_years

```python
import ast
import re

def calculate_player_score(trophies_string, competition_weights):
    try:
        trophies_list = ast.literal_eval(trophies_string)
    except (ValueError, SyntaxError):
        trophies_list = []
    
    scores = [] 
    for trophy in trophies_list:
        match = re.match(r'(\d+)x\s(.+)', trophy)
        if match:
            count = int(match.group(1))
            trophy_name = match.group(2)
            score = competition_weights.get(trophy_name, 0) * count
        else:
            score = competition_weights.get(trophy, 0)
        
        scores.append(score) 
    
    total_score = sum(scores)
    average_score = total_score / len(scores) if scores else 0  
    max_score = max(scores) if scores else 0  
    min_score = min(scores) if scores else 0  
    
    return total_score, average_score, max_score, min_score
```

```python
results = data['trophies_in_recent_years'].apply(lambda x: calculate_player_score(x, competition_weights))
data[['trophies_total_score', 'trophies_average_score', 'trophies_max_score', 'trophies_min_score']] = pd.DataFrame(results.tolist(), index=data.index)

data.drop(columns=['trophies_in_recent_years'], inplace=True)
data
```

```python
def calculate_club_score(trophies_string, competition_weights):
    try:
        trophies_list = ast.literal_eval(trophies_string)
    except (ValueError, SyntaxError):
        trophies_list = []
    
    scores = []
    for trophy in trophies_list:
        match = re.match(r'(\d+)x\s(.+)', trophy)
        if match:
            count = int(match.group(1))
            trophy_name = match.group(2)
            if trophy_name in competition_weights:
                
                score = competition_weights[trophy_name] * count
                scores.append(score)  
        else:
            if trophy in competition_weights:
                scores.append(competition_weights[trophy])
    
    if not scores:  
        min_score_from_weights = min(competition_weights.values()) if competition_weights else 0
        
        return [0, 0, 0, min_score_from_weights]
    
    total_score = sum(scores)
    average_score = total_score / len(scores) if scores else 0
    max_score = max(scores) if scores else 0
    
    min_score = min(scores) if scores else min(competition_weights.values())

    return total_score, average_score, max_score, min_score
```

```python
results = data['club_trophies'].apply(lambda x: calculate_club_score(x, club_competition_weights))

data[['club_trophies_total_score', 'club_trophies_average_score', 'club_trophies_max_score', 'club_trophies_min_score']] = pd.DataFrame(results.tolist(), index=data.index)

data.drop(columns=['club_trophies'], inplace=True)
data
```
