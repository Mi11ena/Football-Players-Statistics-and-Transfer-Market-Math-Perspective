```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('../main.xlsx')
data
```

```python
position_columns = data['position'].str.get_dummies(sep=' - ')
data_expanded = pd.concat([data, position_columns], axis=1)
```

```python
# Create new groups for positions
data_expanded['Defender_Group'] = data_expanded[['Defender', 'Centre-Back', 'Sweeper']].sum(axis=1)
data_expanded['Midfielder_Group'] = data_expanded[['Attacking Midfield', 'Central Midfield', 'Defensive Midfield', 'midfield']].sum(axis=1)
data_expanded['Striker_Group'] = data_expanded[['Attack', 'Centre-Forward', 'Second Striker']].sum(axis=1)
data_expanded['Winger_Group'] = data_expanded[['Left Midfield', 'Left Winger', 'Right Winger', 'Right Midfield', 'Left-Back', 'Right-Back']].sum(axis=1)
```

```python
data_expanded[['Defender_Group', 'Midfielder_Group', 'Winger_Group', 'Striker_Group']]
```

```python
columns_to_drop = [
    'Unnamed: 0', 'place_of_birth', 'Attack', 'Attacking Midfield', 'Central Midfield', 
    'Centre-Back', 'Centre-Forward', 'Defender', 'Defensive Midfield', 'Left Midfield', 
    'Left Winger', 'Left-Back', 'Right Midfield', 'Right Winger', 'Right-Back', 
    'Second Striker', 'Sweeper', 'midfield'
]

data_expanded = data_expanded.drop(columns=columns_to_drop, errors='ignore')
data_expanded
```
